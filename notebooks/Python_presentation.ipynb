```python
# If statement:

for i in range(1,100):    
    if i%3==0 and i%5!=0: x="Fizz"
    elif i%3!=0 and i%5==0: x="Buzz"
    elif i%3==0 and i%5==0: x="Fizz Buzz"
    else: x=i
    print(i,x)
    
# Python uses indentation instead of braces to determine the scope of expressions
# All lines must be indented the same amount to be part of the scope (or indented more if part of an inner scope
```

```python
# While Loop:
x = 1
while x < 10 :
    print (x)
    x = x + 1
```

```python
# For loop
for j in range(1,5):
    print (j)
    
for j in range(5):
    print (j)
```

```python
# break and continue:

# The break statement breaks out of the smallest enclosing for or while loop.

for n in range(2, 10):
    for x in range(2, n):
        if n % x == 0:
            print (n, 'equals', x, '*', n/x)
            break
    else:
        print (n, 'is a prime number')

# The continue statement continues with the next iteration of the loop

for num in range(2, 10):
    if num % 2 == 0:
        print ("Found an even number", num)
        continue
    print ("Found an odd number", num)
```

```python
# Functions
def max(x,y) :
    if x < y :
        return y
    else :
        return x

max_num = max(2,6)     # call function
print ("max number is %d" %max_num)

# Function can be a parameter
def foo(f, a) :
    return f(a)
def bar(p) :
    return p * p
    print(p)

foo(bar,3)

print(p)  #local variables can be accessed only inside the function in which they are declared
```

```python
#Class

# A class is a way to take a grouping of functions and data and place them inside a container
# so you can access them with the . (dot) operator.

class MyClass(object): 
    i = 12345 
    def f(self): 
        return self.i 

MyClass.i
MyClass.f
```
